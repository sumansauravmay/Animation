```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Month": ["Jan","Feb","Mar","Apr","May"],
    "Sales": [12000,15000,17000,14000,19000]
})

plt.plot(df["Month"], df["Sales"])
plt.title("Sales Trend Over Months")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```
df = pd.DataFrame({
    "City": ["Delhi","Mumbai","Bangalore","Kolkata"],
    "Customers": [220,180,140,90]
})

plt.bar(df["City"], df["Customers"])
plt.title("Customer Distribution by City")
plt.xlabel("City")
plt.ylabel("Number of Customers")
plt.show()
```

```
df = pd.DataFrame({
    "Spending": [2000, 3500, 1500, 4000, 2800],
    "Orders": [7, 12, 4, 15, 8]
})
plt.scatter(df["Spending"], df["Orders"])
plt.title("Scatter Plot: Spending vs Orders")
plt.xlabel("Customer Spending")
plt.ylabel("Number of Orders")
plt.show()
```
